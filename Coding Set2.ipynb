Question 1 :

An airline tracks flight delays (in minutes) for 20 flights. Analyze the flight delays to calculate percentiles, detect outliers, and evaluate the overall distribution.
DataSet:
delays = [15, 30, 45, 20, 25, 100, 5, 60, 35, 50, 
          120, 80, 10, 75, 90, 200, 55, 40, 70, 150]
Expected Output:

Percentiles (10th, 25th, 50th, 75th, 90th): [ 14.5   28.75  52.5   82.5  123.  ]
IQR: 53.75
Outliers: [200]


```python
delays = [15, 30, 45, 20, 25, 100, 5, 60, 35, 50, 120, 80, 10, 75, 90, 200, 55, 40, 70, 150]
import numpy as np
print ("Percentiles (10th, 25th, 50th, 75th, 90th):", np.percentile(delays, [10, 25, 50, 75, 90]))
Q1 = np.percentile(delays, 25)
Q3 = np.percentile(delays, 75)
IQR = Q3 - Q1
Rule = 1.5*IQR
LO = Q1 - Rule
GO = Q3 + Rule
Outlier = [x for x in delays if (x>GO) or (x<LO)]
print ("IQR:", IQR)
print("Outliers:", Outlier)

import seaborn as sns 
import matplotlib.pyplot as plt
sns.boxplot(x= delays)
plt.title("Box plot for flight delays")
plt.xlabel("Flight delays (in minutes)")
```

Question 2 :

A company wants to analyze the salary distribution of its employees to understand the central tendency and determine whether the data is skewed.
DataSet:
salaries = [30000, 32000, 35000, 37000, 40000, 42000, 45000, 47000, 50000, 55000,
            60000, 62000, 65000, 67000, 70000, 72000, 75000, 80000, 85000, 90000]
Expected Output:
Mean Salary: 56950.0
Median Salary: 57500.0
Mode Salary: 30000
The data is Left Skewed (Negative Skew)


```python
salaries = [30000, 32000, 35000, 37000, 40000, 42000, 45000, 47000, 50000, 55000, 60000, 62000, 65000, 67000, 70000, 72000, 75000, 80000, 85000, 90000]
from scipy import stats
import numpy as np
mean=np.mean(salaries)
mode=stats.mode(salaries, keepdims=True).mode[0]
print("Mean Salary:",mean)
print("Median Salary:",np.median(salaries))
print("Mode Salary:",mode)

if mean > mode:
    print("The data is Left Skewed (Negative Skew)")
elif mean < mode:
    print("The data is Right Skewed (Positive Skew)")
else:
    print("The data is symmetrical")

import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(salaries, kde=True)
plt.title("Histogram of Employee Salary")
```

Question 4:

A pharmaceutical company conducted a clinical trial with two groups: one receiving medication and the other a placebo. How do you perform a hypothesis test to determine the effectiveness of the medication?
Dataset:
medication_group = [110, 115, 108, 102, 107, 99, 111, 104, 109, 101]
placebo_group = [120, 125, 130, 122, 128, 119, 124, 127, 123, 126]
Expected Output:
T-Statistic: -9.201427649220966
P-Value: 3.163912817600812e-08
Reject the null hypothesis: The medication is effective.


```python
medication_group = [110, 115, 108, 102, 107, 99, 111, 104, 109, 101]
placebo_group = [120, 125, 130, 122, 128, 119, 124, 127, 123, 126]

from scipy.stats import ttest_ind
statistic, Value = ttest_ind(medication_group,placebo_group)
print (f" T-Statistic: {statistic} P-Value: {Value}")

if(Value<=0.05):
    print("Reject the null hypothesis: The medication is effective")
else:
    print("Accept the null hypothesis: The medication is no effective")
```

Question 5 :

A company conducted a customer satisfaction survey where customers rated their experience on a scale of 1 to 10. Analyze the survey results to calculate descriptive statistics and visualize the distribution of customer satisfaction ratings.
Sample DataSet:  
ratings = [8, 9, 7, 5, 6, 10, 9, 4, 7, 8, 
           6, 9, 10, 5, 8, 7, 6, 9, 10, 7]
Expected Output:
Mean Rating: 7.5
Median Rating: 7.5
Mode Rating: 7
Standard Deviation: 1.746424919657298


```python
ratings = [8, 9, 7, 5, 6, 10, 9, 4, 7, 8, 6, 9, 10, 5, 8, 7, 6, 9, 10, 7]
import numpy as np
import scipy.stats as st

print ("Mean Rating:", np.mean(ratings))
print ("Median Rating:", np.median(ratings))
print ("Mode Rating:", st.mode(ratings))
print ("Standard Deviation:", np.std(ratings))

import seaborn as sns
sns.histplot(ratings, kde=True)
plt.title("Distribution of customer satisfaction ratings")
plt.xlabel("Customer Satisfication Ratings")
plt.ylabel("Frequency")
```
