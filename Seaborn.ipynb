```python
import pandas as pd
```

```python
dataset= pd.read_csv("Preprocess_Placement.csv")
dataset.isna().sum()
```

```python
dataset
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

UNIVARIATE

```python
sb.histplot(dataset["ssc_p"],kde=True)
plt.show
```

BIVARIATE

Joint_plot

```python
sb.jointplot(x='ssc_p',y='salary',data= dataset)
plt.show
```

```python
sb.jointplot(x='ssc_p',y='salary',data= dataset, kind="kde")
plt.show
```

```python
sb.jointplot(x='ssc_p',y='salary',data= dataset, kind="hex")
plt.show
```

Pair_plot

```python
sb.pairplot(dataset)
plt.show()
```

```python
sb.pairplot(dataset,hue= "gender")
plt.show()
```

Categorical Value:

```python
sb.stripplot(x="ssc_b", y="ssc_p", data=dataset)
```

```python
sb.swarmplot(x="ssc_b", y="ssc_p", data=dataset)
```
